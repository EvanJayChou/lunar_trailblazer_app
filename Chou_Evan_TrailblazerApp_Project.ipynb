Line plot of Solar Panel Voltage as a function of Time

```python
import matplotlib.pyplot as plt
import pandas as pd

df['Satellite Date/Time UTC'] = pd.to_datetime(df['Satellite Date/Time UTC'])

plt.figure(figsize=(12, 6))
plt.plot(df['Satellite Date/Time UTC'], df['Solar Panel Voltage X mV'], label='X', color='blue')
plt.plot(df['Satellite Date/Time UTC'], df['Solar Panel Voltage Y mV'], label='Y', color='green')
plt.plot(df['Satellite Date/Time UTC'], df['Solar Panel Voltage Z mV'], label='Z', color='red')
plt.xlabel('Time')
plt.ylabel('Feature Value')
plt.title('Solar Panel Voltage Over Time')
plt.legend()
plt.show()
```

Line plot of Boost Converter Temp as a function of Time

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Satellite Date/Time UTC'], df['Boost Converter Temp 1 C'], label='1', color='blue')
plt.plot(df['Satellite Date/Time UTC'], df['Boost Converter Temp 2 C'], label='2', color='green')
plt.plot(df['Satellite Date/Time UTC'], df['Boost Converter Temp 3 C'], label='3', color='red')
plt.xlabel('Time')
plt.ylabel('Feature Value')
plt.title('Boost Converter Temp Over Time')
plt.legend()
plt.show()
```

Line plot of Solar Panel Temperature as a function of Time

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Satellite Date/Time UTC'], df['Solar Panel Temp X+ C'], label='X+', color='blue')
plt.plot(df['Satellite Date/Time UTC'], df['Solar Panel Temp X- C'], label='X-', color='green')
plt.plot(df['Satellite Date/Time UTC'], df['Solar Panel Temp Y+ C'], label='Y+', color='red')
plt.plot(df['Satellite Date/Time UTC'], df['Solar Panel Temp Y- C'], label='Y-', color='yellow')
plt.xlabel('Time')
plt.ylabel('Feature Value')
plt.title('Solar Panel Temperature Over Time')
plt.legend()
plt.show()
```

Line plot of Sun Sensor as a function of time

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Satellite Date/Time UTC'], df['Sun Sensor X+'], label='X+', color='blue')
plt.plot(df['Satellite Date/Time UTC'], df['Sun Sensor Y+'], label='Y+', color='green')
plt.plot(df['Satellite Date/Time UTC'], df['Sun Sensor Z+'], label='Z+', color='red')
plt.xlabel('Time')
plt.ylabel('Feature Value')
plt.title('Sun Sensor Over Time')
plt.legend()
plt.show()
```

There is a high correlation between the 3 features of the datset at specific time frames, therefore, I will analyze the line plots for the average means of each feature and plot them all in one line plot as a function of time. 

```python
voltage_df = df[['Solar Panel Voltage X mV', 'Solar Panel Voltage Y mV', 'Solar Panel Voltage Z mV']].mean(axis=1)
boost_df = df[['Boost Converter Temp 1 C', 'Boost Converter Temp 2 C', 'Boost Converter Temp 3 C']].mean(axis=1)
temp_df = df[['Solar Panel Temp X+ C', 'Solar Panel Temp X- C', 'Solar Panel Temp Y+ C', 'Solar Panel Temp Y- C']].mean(axis=1)

# Performing data manipulation to get a better image of the correlations
voltage_df = (voltage_df - 4000) / 10
boost_df = boost_df * 2


plt.figure(figsize=(12, 6))
plt.plot(df['Satellite Date/Time UTC'], voltage_df, label='Solar Panel Voltage', color='blue')
plt.plot(df['Satellite Date/Time UTC'], boost_df, label='Boost Converter Temp', color='green')
plt.plot(df['Satellite Date/Time UTC'], temp_df, label='Solar Panel Temp', color='red')
plt.xlabel('Time')
plt.ylabel('Feature Value')
plt.title('Mean of Solar Panel Voltage, Boost Converter Temp, and Solar Panel Temp Over Time')
plt.legend()
plt.show()
```
